```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures

import torch
import mctorch.nn as mnn
import mctorch.optim as moptim
from sympy import sin, cos, Matrix, symbols, lambdify
```

```python
# generate dataset.

def getX(ed): # on [0,ed]
    rand = np.random.random()
    term = rand*ed
    return term

#(same function)
def getY(ed): # on [0,ed]
    rand = np.random.random()
    term = rand*ed
    return term

def getZ(x,y):
    term = np.sin(x) - np.cos(y)
    return term

data = []
size = 2**11
for p in range(size):
    x = getX(2*np.pi)
    y = getY(2*np.pi)
    z = getZ(x,y)
    data.extend([[x,y,z]])

data=np.array(data)
```

```python
# Plot the dataset.
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')
ax.scatter(data.transpose()[0],data.transpose()[1],data.transpose()[2],c=data.transpose()[2],cmap="jet")
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
plt.show()
```

```python
def getFeats(p):
    t0 = 1
    t1 = p[0]
    t2 = p[1]
    t3 = p[2]
    t4 = np.cos(p[0])
    t5 = np.cos(p[1])
    t6 = np.cos(p[2])
    t7 = np.sin(p[0])
    t8 = np.sin(p[1])
    t9 = np.sin(p[2])
    row = [t0, t1, t2, t3, t4, t5, t6, t7, t8, t9]
    return row
```

```python
def getFeatMat(data, getFeats):
    B = []
    for p in data:
        B.extend([getFeats(p)])

    B = np.array(B)
    return B
```

```python
B = getFeatMat(data, getFeats)
```

```python
B.shape
```

```python
def tryDim(B, d, criterion=torch.nn.L1Loss(), optimizer=moptim.rAdagrad, lr=1e-2, n_epochs=5000):
    manifold_param = mnn.Parameter(manifold=mnn.Stiefel(B.shape[1],d))
    Btensor = torch.tensor(B).float() #extB.detach()
    
    def model(mat):
        term1 = torch.matmul(Btensor, mat.float())
        return term1
    
    y = torch.zeros((B.shape[0],d))
    optimizer = optimizer(params = [manifold_param], lr=lr)
    
    for epoch in range(n_epochs):
        y_pred = model(manifold_param)
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
    return loss.item(), manifold_param
```

```python
t1 = tryDim(B, 1)
t1
```

```python
def getExtendedFeatureMatrix2(B,J,dim):
    n = B.shape[0]
    m = B.shape[1]
    ExtArray = []
    for j in range(n):
        newBthing = np.append(B[j],np.zeros((dim-1)*m))
        Barray = np.array([np.roll(newBthing,i*m) for i in range(dim)])
        Barray = torch.tensor(Barray).float()
        thing = torch.matmul(J[j], Barray)
        #thing = np.dot(J[j],Barray)
        ExtArray.extend(thing)
        
    ExtArray = torch.stack(ExtArray, dim=0)
    #ExtArray = np.array(ExtArray)
    return ExtArray
```

```python
extB = getExtendedFeatureMatrix2(B2,J,2)
```

```python
def tryDimV(extB, d, n_epochs=1000, optimizer=moptim.rSGD, lr=1e-2, criterion=torch.nn.MSELoss()):
    manifold_param = mnn.Parameter(manifold=mnn.Stiefel(extB.shape[1],d))
    Btensor = extB.detach() #torch.tensor(B).float()
    
    def model(mat):
        term1 = torch.matmul(Btensor, mat)
        #term2 = torch.norm(term1)
        #return term2
        return term1
    
    y = torch.zeros((B.shape[0],d))
    
    #optimizer = moptim.rAdagrad(params = [manifold_param], lr=1e-2)
    optimizer = optimizer(params = [manifold_param], lr=lr) #moptim.rSGD(params = [manifold_param], lr=1e-2)
    #criterion = torch.nn.MSELoss() #torch.nn.L1Loss() #torch.nn.MSELoss()
    for epoch in range(n_epochs):
        y_pred = model(manifold_param)
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
    return loss.item(), manifold_param
```
